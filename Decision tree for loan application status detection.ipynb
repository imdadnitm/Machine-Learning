```python
prd_df=df.drop("Average_Number_of_Dependents",axis=1)
```

```python
prd_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics 
```

```python
x=df[["Gender","Married","Average_Number_of_Dependents","Self_Employed","Credit_History","Property_Area","Income","Education"]]
```

```python
x["Gender"]=x["Gender"].apply(lambda g:0 if g=="Male" else 1)
x["Married"]=x["Married"].apply(lambda g:0 if g=="No" else 1)
x["Self_Employed"]=x["Self_Employed"].apply(lambda g:0 if g=="No" else 1)
x["Income"]=x["Income"].apply(lambda g:1 if g=="high" else 0)
x["Property_Area"]=x["Property_Area"].apply(lambda g:0 if g=="Rural" else 1)
x["Education"]=x["Education"].apply(lambda g:1 if g=="Graduate" else 0)
```

```python
y=prd_df["Application_Status"].apply(lambda g:0 if g=="N" else 1)
#y["Application_Status"]=y["Application_Status"]
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=32)
```

```python
pred_model=DecisionTreeClassifier()
```

```python
pred_model.fit(x_train,y_train)
```

```python
y_predict=pred_model.predict(x_test)
```

```python
acc=metrics.accuracy_score(y_test, y_predict)
```

```python
print("Accuracy of the model: ",acc)
```
